```python
# import matplotlib.pyplot library 
import matplotlib.pyplot as plt 

data = [16, 25, 47, 56, 23, 45, 19, 55, 44, 27] 

# separating the stem parts 
stems = [1, 1, 2, 2, 2, 4, 4, 4, 5, 5] 

plt.ylabel('Data') # for label at y-axis 

plt.xlabel('stems') # for label at x-axis 

plt.xlim(0, 10) # limit of the values at x axis 

plt.stem(stems, data,use_line_collection=True) # required plot 
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')

dataset.head()
```

```python
sns.distplot(dataset['fare'])
```

```python
sns.jointplot(x='age', y='fare', data=dataset)
```

```python
sns.jointplot(x='age', y='fare', data=dataset, kind='hex')
```

```python
sns.pairplot(dataset, hue='sex')
```

```python
sns.stripplot(x='sex', y='age', data=dataset)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived', split=True)
```
